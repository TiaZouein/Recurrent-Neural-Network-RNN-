## Recurrent Neural Network

Les réseaux neuronaux récurrents (RNN) sont l'algorithme de pointe pour les données séquentielles. **Ils sont utilisés par Siri d'Apple et la recherche vocale de Google.**

Comme de nombreux autres algorithmes d'apprentissage profond, les réseaux neuronaux récurrents sont relativement anciens. Ils ont été initialement créés dans les années 1980, mais ce n'est que ces dernières années que nous avons vu leur véritable potentiel. 

Pour comprendre correctement les RNN, il faut avoir une connaissance pratique des feed-forward neural network "normaux". Ces réseaux sont principalement utilisés pour la reconnaissance de motifs et peuvent être illustrés comme suit

```python
from IPython.display import Image, display

image_path1 = "ffnn.png"
image_path2 = "ffnn2.png"

width = 300
height = 300

display(Image(filename=image_path1, width=width, height=height), Image(filename=image_path2, width=width, height=height))
```

Dans un feed-forward neural network, l'information ne se déplace que dans une direction, de la couche d'entrée, à travers les couches cachées, jusqu'à la couche de sortie. L'information circule directement à travers le réseau.
Les réseaux neuronaux feed-forward n'ont aucune mémoire de l'entrée qu'ils reçoivent et sont mauvais pour prédire ce qui va suivre. Parce qu'un feed-forward neural network ne considère que l'entrée actuelle, il n'a aucune notion d'ordre dans le temps. Il ne peut tout simplement pas se souvenir de ce qui s'est passé dans le passé, sauf pendant son entraînement.

Les données séquentielles sont essentiellement des données ordonnées dans lesquelles des éléments liés se suivent. Les exemples incluent les données financières ou la séquence d'ADN. Le type le plus populaire de données séquentielles est peut-être les données de séries temporelles, qui ne sont qu'une série de points de données répertoriés dans l'ordre du temps.

**Dans un RNN, l'information circule à travers une boucle. Lorsqu'il prend une décision, il considère l'entrée actuelle ainsi que ce qu'il a appris des entrées qu'il a reçues précédemment.**

En raison de leur mémoire interne, les RNN peuvent se souvenir de choses importantes sur l'entrée qu'ils ont reçue, ce qui leur permet d'être très précis dans la prédiction de ce qui va suivre. C'est pourquoi ils sont l'algorithme préféré pour les données séquentielles telles que les séries temporelles, la parole, le texte, les données financières, l'audio, la vidéo, la météo et bien plus encore. Les réseaux neuronaux récurrents peuvent former une compréhension beaucoup plus profonde d'une séquence et de son contexte par rapport à d'autres algorithmes.

Une autre bonne façon d'illustrer le concept de la mémoire d'un réseau neuronal récurrent est de l'expliquer avec un exemple : Imaginons qu'on a un feed-forward neural network normal et qu'on lui donne le mot "neurone" en entrée, et qu'il traite le mot caractère par caractère. Au moment où il atteint le caractère "r", il a déjà oublié "n", "e" et "u", ce qui rend presque impossible pour ce type de réseau neuronal de prédire quel caractère viendrait ensuite.

Un réseau neuronal récurrent, en revanche, est capable de se souvenir de ces caractères grâce à sa mémoire interne. Il produit une sortie, copie cette sortie et la boucle à nouveau dans le réseau.

Dans les réseaux neuronaux récurrents, l'architecture récurrente permet aux données de revenir à la couche d'entrée. Cela signifie que les données ne sont pas limitées à une direction à propagation avant. Plus précisément, **dans une couche cachée d'un RNN, la sortie du point temporel précédent deviendra une partie de l'entrée pour le point temporel actuel**. ## Exemples concrets d'utilisation des RNNs: 

**Speech Recognition:**

Les réseaux neuronaux récurrents (RNN) sont largement utilisés dans la reconnaissance vocale en raison de leur capacité à traiter des données séquentielles telles que le langage parlé. Dans les systèmes de reconnaissance vocale, les RNN peuvent analyser et comprendre les modèles temporels et séquentiels présents dans les signaux audio pour convertir la parole humaine en texte écrit. Cette capacité à prendre en compte le contexte et les relations temporelles entre les phonèmes ou les mots rend les RNN particulièrement efficaces pour la reconnaissance précise de la parole dans différentes langues et accents.

**Traduction:**

Les réseaux neuronaux récurrents (RNN) sont également largement utilisés dans le domaine de la traduction automatique. Ces réseaux sont bien adaptés à la traduction en raison de leur capacité à gérer des séquences d'entrée et de sortie de longueurs variables. Dans le contexte de la traduction, un RNN peut prendre une séquence de mots dans une langue source en tant qu'entrée, puis générer une séquence de mots dans une langue cible en sortie. Cette capacité à tenir compte du contexte et de la structure séquentielle des phrases rend les RNN efficaces pour capturer les nuances des langues naturelles.

**Série temporelle:**

Les RNN sont particulièrement adaptés à la modélisation des séries temporelles en raison de leur capacité à capturer les dépendances temporelles complexes et à mémoriser les motifs dans les données séquentielles. En utilisant des couches récurrentes, les RNN peuvent analyser les données passées pour prédire les valeurs futures d'une série temporelle.
Ils sont souvent utilisés dans des applications telles que la prévision financière, la prédiction de la demande, la prévision météorologique... Leur capacité à traiter des données séquentielles leur permet de capturer les tendances, les saisons, les cycles et d'autres motifs temporels complexes présents dans les séries temporelles, ce qui en fait des outils puissants pour l'analyse et la prédiction des données temporelles.
